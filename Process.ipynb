## 1. Data Cleaning
- Data verification such as missing, unfilled, or incorrect entries.
- Replacement or addition. If there is any missing data, it shall be filled using the average method.

```python
File = './Results/unemployment.csv'
```

```python
dt = pd.read_csv(File)
dt
```

```python
# 1. Data Validation
def validate_data(df):
    missing_data = df.isnull().sum()
    print("Number of missing values in the dataset:\n", missing_data)
    return missing_data

# 2. Handling Missing Data
def handle_missing_data(df):
    # Remove rows with missing values if their number is small
    df_cleaned = df.fillna(0)
    return df_cleaned

# 3. Reporting Missing Data
def report_missing_data(missing_data):
    for column, count in missing_data.items():
        if count > 0:
            print(f"The column '{column}' has {count} missing values.")

# Calling the above functions
missing_data = validate_data(dt)
df_cleaned = handle_missing_data(dt)
report_missing_data(missing_data)

output_file = './Results/Clear_data_unemployment.csv'
df_cleaned.to_csv(output_file, index=False)

print(f'df_save stored in {output_file}')

print("\nDataFrame after handling missing data:")
df_cleaned
```
